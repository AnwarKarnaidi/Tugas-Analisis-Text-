```python
# Memproses Data

from transformers import AutoTokenizer # Mengimpor kelas AutoTokenizer dari pustaka transformers

# Pilih model tokenizer
model_names = ['bert-base-uncased', 'roberta-base', 'distilbert-base-uncased'] # daftar nama model tokenizer yang akan digunakan

# kamus yang memetakan nama model ke objek tokenizer yang dimuat menggunakan metode from_pretrained
tokenizers = {name: AutoTokenizer.from_pretrained(name) for name in model_names}


# Tokenisasi dataset
def tokenize_function(examples, tokenizer): # fungsi yang mengambil input teks dari dataset dan menerapkannya ke tokenizer yang diberikan

# mengembalikan teks yang telah ditokenisasi dengan padding hingga panjang maksimal (max_length) dan melakukan truncation jika panjang teks melebihi panjang maksimal
    return tokenizer(examples['text'], padding='max_length', truncation=True)

# Tokenisasi untuk setiap model
tokenized_datasets = {name: dataset.map(lambda x: tokenize_function(x, tokenizer), batched=True) for name, tokenizer in tokenizers.items()}
```

```python
# Memastikan bahwa kolom yang digunakan adalah benar
def tokenize_function(examples, tokenizer):
    return tokenizer(examples['text'], padding='max_length', truncation=True)

# Tokenisasi untuk setiap model
tokenized_datasets = {}  # Inisialisasi kamus kosong untuk menyimpan dataset yang telah ditokenisasi untuk setiap model
for name, tokenizer in tokenizers.items(): # Melakukan iterasi pada setiap item dalam kamus tokenizers yang berisi nama model dan objek tokenizer
    tokenized_dataset = dataset.map(lambda x: {'input_ids': tokenizer(x['text'], padding='max_length', truncation=True)['input_ids'],
                                               'attention_mask': tokenizer(x['text'], padding='max_length', truncation=True)['attention_mask'],
                                               'labels': x['coarse_label']}, batched=True)
# Menggunakan metode map pada dataset untuk menerapkan fungsi lambda yang memanggil tokenizer pada setiap teks (x['text'])
# Menghasilkan input_ids dan attention_mask dari tokenizer, serta menambahkan label (x['coarse_label']) ke dalam kamus hasil
# Parameter batched=True digunakan untuk memproses data dalam batch, yang meningkatkan efisiensi pemrosesan

    tokenized_datasets[name] = tokenized_dataset # Menyimpan hasil dataset yang telah ditokenisasi ke dalam kamus tokenized_datasets dengan nama model sebagai kunci

```

```python
from transformers import DataCollatorWithPadding
from torch.utils.data import DataLoader

# Menggunakan data collator
data_collators = {name: DataCollatorWithPadding(tokenizer=tokenizer) for name, tokenizer in tokenizers.items()}

# DataLoader
train_loaders = {
    name: DataLoader(tokenized_datasets[name]['train'], batch_size=8, shuffle=True, collate_fn=data_collators[name]) for name in model_names
}
test_loaders = {
    name: DataLoader(tokenized_datasets[name]['test'], batch_size=8, shuffle=False, collate_fn=data_collators[name]) for name in model_names
}

```

```python
# Fine-Tuning Model

from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
# AutoModelForSequenceClassification: Kelas ini digunakan untuk memuat model pre-trained yang dapat digunakan untuk tugas klasifikasi teks.
# Trainer: Kelas ini memudahkan proses pelatihan dan evaluasi model dengan menyediakan antarmuka untuk melatih model menggunakan data yang disediakan.
# TrainingArguments: Kelas ini digunakan untuk mengonfigurasi berbagai parameter pelatihan seperti batch size, learning rate, dan jumlah epoch.
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
# accuracy_score dan precision_recall_fscore_support: Fungsi dari sklearn.metrics untuk menghitung metrik evaluasi seperti akurasi, presisi, recall, dan F1-score.

# Fungsi evaluasi metrik
def compute_metrics(p): # Fungsi ini digunakan untuk menghitung metrik evaluasi dari hasil prediksi model.
    pred, labels = p
    pred = np.argmax(pred, axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, pred, average='macro')
    acc = accuracy_score(labels, pred)
    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}

# Fine-tuning
trainers = {}
for name in model_names:

  # Memuat model pre-trained untuk tugas klasifikasi dengan 6 label.
    model = AutoModelForSequenceClassification.from_pretrained(name, num_labels=6)

    training_args = TrainingArguments(
        output_dir=f'./results/{name}',
        eval_strategy='epoch',
        learning_rate=2e-5,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        num_train_epochs=3,
        weight_decay=0.01,
        logging_dir='./logs',
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_datasets[name]['train'],
        eval_dataset=tokenized_datasets[name]['test'],
        tokenizer=tokenizers[name],
        compute_metrics=compute_metrics
    )

    trainer.train()
    trainers[name] = trainer
```

```python
# Evaluasi Model

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def evaluate_model(trainer, test_dataset):
    results = trainer.predict(test_dataset)
    pred_labels = np.argmax(results.predictions, axis=1)
    true_labels = results.label_ids

    acc = accuracy_score(true_labels, pred_labels)
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, pred_labels, average='macro')

    print(f"Accuracy: {acc}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")

    cm = confusion_matrix(true_labels, pred_labels)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Evaluasi model
for name in model_names:
    print(f"Evaluasi model {name}")
    evaluate_model(trainers[name], tokenized_datasets[name]['test'])
```

```python
# Analisis Hasil

import pandas as pd

for name in model_names:
    logs = pd.DataFrame(trainers[name].state.log_history)
    plt.plot(logs['epoch'], logs['loss'], label=f'{name} - Training Loss')
    plt.plot(logs['epoch'], logs['eval_loss'], label=f'{name} - Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
